```python
print(a3)
```

```python
print(a3[1,0,1]) # a3[dim,row,com]
```

```python
# slicing
print(a1)
```

```python
print(a1[2:5])
```

```python
a2
```

```python
print(a2[0,:])
print(a2[:,2]) # a2[row,col(':' means all)]
print(a2[1,::3])
```

# Stacking

```python
# Example of Horizontal Stacking
a = np.arange(1, 5).reshape(2, 2)
b = np.arange(5, 9).reshape(2, 2)

print('Array A:\n', a)
print('Array B:\n', b)

print('\nHorizontal Stack:')
print(np.hstack((a, b)))
```

```python
# Example of Vertical Stacking
print('Vertical Stack:')
print(np.vstack((a, b)))
```

# Splitting

```python
import numpy as np
# Create a 4x4 array for splitting
a = np.arange(16).reshape(4,4)
print('Original Array:\n', a)

# Horizontal splitting (splitting columns into equal parts)
# This splits the 4x4 into two 4x2 arrays
print('\nHorizontal Split (into 2):')
print(np.hsplit(a, 2))
```

```python
import numpy as np
# Vertical splitting (splitting rows into equal parts)
# This splits the 4x4 into two 2x4 arrays
print('Vertical Split (into 2):')
print(np.vsplit(a, 2))
```
