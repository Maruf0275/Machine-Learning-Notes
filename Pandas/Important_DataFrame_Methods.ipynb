### Important DataFrame Methods

```python
import numpy as np
import pandas as pd
ipl = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/ipl-matches.csv')
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/movies.csv')
```

```python
# value_counts(series and dataframe)
marks = pd.DataFrame(
    [
       [100,80,10],
       [90,70,7],
       [120,100,14],
       [80,70,14],
       [80,70,14]
    ],columns = ['iq','marks','package']
)
marks.value_counts()
```

```python
# find which player has won most potm -> in finals and qualifiers
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
# Toss decision plot
ipl['TossDecision'].value_counts().plot(kind = 'pie')
```

```python
# how many matches each team has played
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending = False)
```

```python
# sort_values(series and dataframe) -> ascending -> na_position -> inplace -> mutiple cols
movies.sort_values('title_x')
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

students
```

```python
students.sort_values('name',na_position='first')
```

```python
# multiple cols
movies.sort_values(['year_of_release','title_x'])
```

```python
batsman = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/batsman_runs_ipl.csv')
batsman
```

```python
# rank(series)
batsman['batting_rank'] = batsman['batsman_run'].rank(ascending=False)
batsman.sort_values('batting_rank').head(5)
```

```python
# sort_index(series and dataframe)
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks)
marks_series.sort_index()
```

```python
movies.sort_index(ascending=False)
```

```python
# set_index(dataframe)
batsman.set_index('batter',inplace=True)
```

```python
# reset_index(series and dataframe) -> drop parameter
batsman.reset_index()
```

```python
# how to replace existing index without loosing
batsman.reset_index().set_index('batting_rank')
```

```python
# series to dataframe using reset_index
marks_series.reset_index() # it will be convert to dataframe
```
