### How to use groupby in pandas

```python
import pandas as pd

data = {
    'City': ['Dhaka', 'Dhaka', 'Chittagong', 'Chittagong'],
    'Salary': [50000, 60000, 40000, 45000]
}

df = pd.DataFrame(data)
df
```

```python
# groupby
cities = df.groupby('City')
```

```python
df.groupby('City')['Salary'].mean()
```

```python
len(cities)
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/imdb-top-1000.csv')
movies.head(1)
```

```python
genres = movies.groupby('Genre')
```

```python
# aggregation
genres.sum()
```

```python
genres.min()
```

```python
genres.max()
```

```python
genres.agg({'IMDB_Rating':['mean','max']})
```

```python
# find the top 3 genres by total earning
genres['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest avg IMDB rating
genres['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find the director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find number of movies by each actor
movies['Star1'].value_counts()

movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
movies.head(1)
```

```python
# agg method
genres.agg({'IMDB_Rating':['mean','max'],
            'Runtime':['mean'],
            'Gross':['max','min','mean']})
```

```python
# find the highest rated movie in each genre
movies.sort_values('IMDB_Rating', ascending=False).groupby('Genre').head(1)
```

```python
# groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
# size
duo.size()
# get_group
duo.get_group(('Aamir Khan','Amole Gupte'))
duo.get_group(('Christopher Nolan','Christian Bale'))
```

```python
# find the most earning actor and director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# agg on multiple groupby - selecting only numeric columns to avoid TypeError
duo[['IMDB_Rating', 'Gross']].agg(['min', 'max', 'mean'])
```
