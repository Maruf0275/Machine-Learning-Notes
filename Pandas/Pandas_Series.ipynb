### Introduction to Pandas
Pandas is the standard tool for data manipulation in Python.

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Series from lists

```python
# String
country = ['Bangladesh','India','USA','Pakistan','Nepal']
print(pd.Series(country))
animal = ['Tiger','Lion','Elephant','Cat','Dog']
print(pd.Series(animal))
```

```python
# integers
runs = [18,39,49,49,100]
print(pd.Series(runs))
```

```python
# Custom Index
marks = [50,66,73,80]
subjects = ['Maths','English','Science','Bangla']
print(pd.Series(marks, index = subjects))
```

```python
# Setting a name
marks = [50,66,73,80]
subjects = ['Maths','English','Science','Bangla']

print(pd.Series(marks, index = subjects,name = 'Marks of Maruf'))
```

### Series from dictionary

```python
marks = {
    'Maths':50,
    'English':66,
    'Science':73,
    'Bangla':80
}

print(pd.Series(marks,name = 'Marks of Maruf'))
```

### Series Attribute

```python
# size
marks = {
    'Maths':50,
    'English':66,
    'Science':73,
    'Bangla':80
}
marks_series = pd.Series(marks,name = 'Marks fo Maruf')
print(marks_series.size)
```

```python
# dtype
marks_series.dtype
```

```python
# name
marks_series.name
```

```python
# is_unique
print(marks_series.is_unique)

pd.Series([1,1,3,4,5]).is_unique
```

```python
# index
marks_series.index # Store values in index
```

```python
# values
marks_series.values
```
